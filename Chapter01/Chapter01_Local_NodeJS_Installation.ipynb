# Chapter 01: Local Installation via Node.js (Windows/Mac/Linux)

## Introduction
The simplest way to get started with n8n is a local installation using Node.js. This method is ideal for testing and development on your own computer.

## Objective
Install n8n locally using the Node.js runtime environment for development and testing purposes across multiple operating systems.

## Tools & Technologies
- Node.js
- npm
- nvm
- Yarn/pnpm

## Required Steps
1. **Node.js Installation**: Ensure you have Node.js 18+ installed.
2. **n8n Installation**: Running the global install command.

## Commands
```bash
# Install n8n globally
npm install n8n -g

# Start n8n
n8n start
```

## Conclusion
You have successfully installed n8n locally. You can access the UI at `http://localhost:5678`.
