# Chapter 03: PM2 Process Management

## Introduction
PM2 is a production process manager for Node.js applications. It ensures n8n restarts automatically if it crashes or if the server reboots.

## Commands
```bash
# Install PM2
npm install pm2 -g

# Start n8n with PM2
pm2 start n8n -- start

# Save process list for reboot persistence
pm2 save
pm2 startup
```

## Conclusion
n8n is now running as a background service.
