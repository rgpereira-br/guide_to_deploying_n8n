# Chapter 03: PM2 Process Management

## Introduction
PM2 is a production process manager for Node.js applications and other services.

## Objective
Configure PM2 process manager to run n8n as a resilient background service with auto-restart.

## Tools & Technologies
- PM2
- Node.js

## Commands
```bash
# Install PM2
npm install pm2 -g

# Start n8n with PM2
pm2 start n8n -- start
```
