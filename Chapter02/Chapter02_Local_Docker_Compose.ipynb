# Chapter 02: n8n in Local Docker Compose

## Introduction
Docker provides a consistent environment and makes it easy to keep your data persistent even when the container is restarted.

## Required Steps
1. **Docker Desktop**: Install on Windows/Mac/Linux.
2. **docker-compose.yml**: Create a configuration file.
3. **Volumes**: Map a local folder to `/home/node/.n8n`.

## Configuration
```yaml
services:
  n8n:
    image: n8nio/n8n
    restart: always
    ports:
      - "5678:5678"
    volumes:
      - ./n8n_data:/home/node/.n8n
```

## Conclusion
Your n8n instance is now containerized and persistent.
