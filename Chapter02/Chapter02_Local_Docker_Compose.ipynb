# Chapter 02: Running n8n in Local Docker Compose

## Introduction
Docker provides a consistent environment and makes it easy to keep your data persistent even when the container is restarted.

## Objective
Deploy n8n in a containerized environment using Docker Compose to ensure environment consistency and persistent storage for local workflow data.

## Tools & Technologies
- Docker
- Docker Compose
- Docker Desktop

## Configuration Example
```yaml
services:
  n8n:
    image: n8nio/n8n
    restart: always
    ports:
      - "5678:5678"
    volumes:
      - ./n8n_data:/home/node/.n8n
```

## Conclusion
Your n8n instance is now containerized and persistent.
