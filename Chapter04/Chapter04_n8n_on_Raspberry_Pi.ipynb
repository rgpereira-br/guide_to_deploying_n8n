# Chapter 04: n8n on Raspberry Pi

## Introduction
A Raspberry Pi is an excellent low-power option for a 24/7 automation hub.

## Required Steps
1. Use Raspberry Pi OS (Lite recommended).
2. Increase Swap space if using a 1GB/2GB model.
3. Install Node.js or Docker.

## Conclusion
Your n8n instance is now running on a dedicated local hardware.
